```python
def categorize_age(age):
    if age <= 10:
        return '0-10'
    elif age <= 20:
        return '10-20'
    elif age <= 30:
        return '20-30'
    elif age <= 40:
        return '30-40'
    elif age <= 50:
        return '40-50'
    else:
        return '50+'
df['Age'] = df['Age'].apply(categorize_age)
sns.set(style="whitegrid")
sns.countplot(x='Age', data=df, hue='Sex')
```

OBSERVATION 2

 Above graph is of Age and Sex. We can see in above graph most people are of
age of 50+


```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings


warnings.filterwarnings("ignore")
sns.set_style("whitegrid")

sns.FacetGrid(df, hue="Sex", height=5.0).map(sns.distplot, "Pclass", bins=[1, 2, 3]).add_legend()
plt.title("Sex and Pclass")
plt.show()
```

```python

average_fare_by_pclass = df.groupby('Pclass')['Fare'].mean()
plt.plot(average_fare_by_pclass.index, average_fare_by_pclass.values)
plt.xlabel('Pclass')
plt.ylabel('Average Fare')
plt.title('Average Fare by Pclass')
plt.show()
```

```python
average_fare_by_cabin = df.groupby('Cabin')['Fare'].mean()
plt.plot(average_fare_by_cabin.index, average_fare_by_cabin.values)
plt.xlabel('Cabin')
plt.ylabel('Average Fare')
plt.title('Average Fare by Cabin')
plt.show()
```

OBSERVATION 6

 Above graph is Average Fare by Cabin. We can see that Cabin B has maximum
